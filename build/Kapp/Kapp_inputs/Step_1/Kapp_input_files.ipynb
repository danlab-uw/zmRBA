```python
import pandas as pd
from pathlib import Path
import sys 

sys.path.append('../../../') 
from custom_functions_ZMRBA import *


input_file = Path('../../../ZMRBA_outputs/RBA_Stoichiometry.xlsx')

output_dir = Path('Kapp_Step1_inputs')
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
df_eqn = pd.read_excel(input_file)
```

```python
sij_pro_and_enz = set()

for i in df_eqn.index:
    if df_eqn.type[i] == 'metabolic':
        continue

    if df_eqn.type[i] == 'proteinWaste':
        continue

    if df_eqn.type[i] == 'biomass':
        continue
    
    if df_eqn.type[i] == 'ribosome':
        continue
    
    if df_eqn.type[i] == 'protein':
        continue

    else: 
        x = metabolites_dict_from_reaction_equation_RBA(df_eqn.reaction[i])
        met_dict = dict()
        for k,v in x.items():
            if k == '':
                continue
            

            try:
                v_float = float(v)
            except ValueError:
                continue

            if v_float.is_integer():
                met_dict[k] = int(v_float)
            else:
                met_dict[k] = v_float

    for k,v in met_dict.items():
        sij_pro_and_enz.add("'" + k + "'.'" + df_eqn.id[i] + "'" + str(v))

    for k in met_dict:
        if k[:4] == 'PRO-':
            sij_pro_and_enz.add("'PRO-" + k[4:] + "'.'PROSYN-" + k[4:] + "'" + '1')

for i in df_eqn.index:
    if df_eqn.id[i] != 'RIBOSYN-ribonuc':
        continue
    else:
        x = metabolites_dict_from_reaction_equation_RBA(df_eqn.reaction[i])
        for k in x:
            if k[:4] == 'PRO-':
                sij_pro_and_enz.add("'PRO-" + k[4:] + "'.'PROSYN-" + k[4:] + "'" + '1')


sij_pro_and_enz = ['/'] + sorted(list(sij_pro_and_enz)) + ['/']

sij_proenz = output_dir / 'sij_pro_and_enz.txt'
with open(sij_proenz, 'w') as f:
    f.write('\n'.join(sij_pro_and_enz))
```
